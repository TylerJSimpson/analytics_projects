```python
#Import libraries

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib

import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

#Import data

df = pd.read_csv(r'C:\Users\Tyler Simpson\Downloads\movies.csv')
```

```python
#Explore data

df.head()
```

```python
#Check for null

for col in df.columns:
    prct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, prct_missing))
```

```python
#Remove null

df = df.dropna(how='any',axis=0) 
```

```python
#Column data types

df.dtypes
```

```python
#Clean columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
#Create correct year column

df['yearcorrect'] = df['released'].astype(str).str.split(', ').str[-1].astype(str).str[:4]
```

```python
#Explore data further

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#Drop duplicates

df.drop_duplicates().head()
```

```python
#Data cleaned, form hypotheses

#Hypothesis 1: budget and gross have a high correlation
#Hypothesis 2: company and gross have a high correlation
```

```python
#Hypothesis 1
#Scatter plot budget vs gross

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('budget vs gross')
plt.xlabel('budget')
plt.ylabel('gross')
plt.show
```

```python
#Hypothesis 1
#Regression plot budget vs gross

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "black"}, line_kws={"color": "blue"})
```

```python
#Hypothesis 1
#Correlation check

#Pearson by default, other options: kendall, spearman
#df.corr(method='pearson')

df.corr()
```

```python
#Hypothesis 1
#Visualize numerical correlation with heatmap

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation matrix for numeric data')
plt.xlabel('Movie features')
plt.ylabel('Movie features')

plt.show
```

```python
#Hypothesis 1 correct
#High correlation between budget and gross
```

```python
#Hypothesis 2
#

df_numerical = df

for col_name in df_numerical.columns:
    if(df_numerical[col_name].dtype == 'object'):
        df_numerical[col_name] = df_numerical[col_name].astype('category')
        df_numerical[col_name] = df_numerical[col_name].cat.codes
        
df_numerical.head()
```

```python
#Hypothesis 2
#Visualize numerical correlation with heatmap

correlation_matrix = df_numerical.corr()

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation matrix for numeric data')
plt.xlabel('Movie features')
plt.ylabel('Movie features')

plt.show
```

```python
#Hypothesis 2
#Unstack correlation

correlation_matrix = df_numerical.corr()
correlation_pairs = correlation_matrix.unstack()
sorted_pairs = correlation_pairs.sort_values()
sorted_pairs
```

```python
#Hypothesis 2
#Unstacked sorted

high_correlation = sorted_pairs[(sorted_pairs) > 0.4]
high_correlation
```

```python
#Hypothesis 2 incorrect
#Low correlation between company and gross

#Next highest significance beyond budget and gross is votes and gross
```
